### Overview

When building a web crawler, it is important to keep in mind the freshness of each page in the crawl index. In an ideal world, pages in the index would be continually refreshed ensuring indexed content mirrors the present web. However, contraints on time, bandwidth and disk space force practical limits on the frequency with which content can be refreshed. With this in mind, let's build an optimal page refresh strategity. The below derivation of one such solution is based on [the work of Olston and Pandey](http://infolab.stanford.edu/~olston/publications/www08.pdf).



### Problem

Let's formalize the fresheness problem. Given two versions of a site: $P$ the currect page and $P'$ the indexed page, a simility measure $D(P,P')$, called the page divergence, can be formulated. The page divergence can, in theory, be formed as any [monotone function](http://en.wikipedia.org/wiki/Monotonic_function). For subsequent calculations we use the classic Jaccard similarity.

$$
D(P,P') = \frac{|F(P)\cup F(P')|}{|F(P)\cap F(P')|}
$$

Here, F(P) referres to the [w-shingles](http://en.wikipedia.org/wiki/W-shingling) of page, P. The Jaccard metric measures the number of K-grams (w-shingles) in common between pages P and P', normalizing by the total number of unique K-grams. Splitting the document into such features allows for a finer resolution metric than the common, naive, binary metric "has any element of a page changed."

Given that we've defined a pairwise page metric, let's now formulate the *optimal* page refresh function. Rephrased, for a fixed number of page refreshses, let's formulate the optimal re-crawl schedule. It can be shown that for any monotonically increasing divergence function, D, the following *utility function* is infact optimal.

$$
U_P(t) = t * D_P(t) - \int_0^t{D_p(t')dt'}
$$

This function seems complex but it's intuitive. Breaking it down piece by piece, $U_P(t)$ is the Utility of refreshing page P at time t given a global, fixed resource limit, $T$. In practice, a page will be be re-crawled when $U_P(t) = T$. This utility is inversely proportional to the intergral of D_P(t) over all of observed time. This integral can be thought of as the net change in page P. Therefore, the more a page changes, the smaller the utility will be and the more frequently the page will be refreshed. Finally, the constant $t * D_P(t)$ is added to $U$ to rescale.

Calculating $U_P(t)$ requires two stages, offline learning and online evaluation.

### Offline learning

Let's assume a crawler has indexed the same page several times over the period of a few weeks. To evaluate utility we must first estimate $D_P(t)$. Everytime a page is re-crawled, the Jaccard distance is calculated between that page and all previous crawls of the page. This requires storing the distance between all pairwise combinations of page instances. This is stored as a list of tuples. Let's say we have 4 versions of a page in the index $P_1, P_2, P_3, P_4$ indexed uniformily on days 1, 2, 3, 4. We can write the tuples (time, distance) as:


$$
[(1,\ D(P_1,P_2)),\ (1,\ D(P_2,P_3)),\ (1,\ D(P_3, P_4)),\\
\ (2,\ D(P_1,P_3)),\ (2,\ D(P_2,P_4)),\ (3,\ D(P_1,P_4))]
$$

$D_P(t)$ is taken as the average $D(P_i,P_{i+t})$. For example, given the above tuples, $D_P(2)=\frac{D(P_1,P_3)+D(P_2,P_4)}{2}$. As another example, $D$ is calculated using a sequence of test pages generated across five days:

Now that we have defined a discrete distribution, $D_P(t)$, we must integrate it to evaluate $U_P(t)$. However, being a discrete distribution, we have to map $D$ to a continuous function before integration. The naive mapping is called linear interpolation (i.e., connect-the-dots). Such a fit would look like the dashed line in the figure above. While this gives a nice fit to $D$, it breaks the principle stated before that utility is only maximized under a monotone ('ever increasing') function. Furthermore, it is likely that linear interpolation is overfitting the divergenece of this exact page. To fix this problem, we *approximate* a continuous $D_P(t)$ by fitting two separate Poisson-like distributions and choosing the better fit. The two equations are written below:

$$
D_P^1(t) \approx 1 - e^{-\lambda_R*t} \\ 
D_P^2(t) \approx 1 - \sum_{i=0}^{K-1}\frac{(\lambda_R*t)^i*e^{-\lambda_R*t}}{i!}
$$

$D_P^1(t)$ describes page content as a "churn" model wherein page content is overwritten at a predictible, average rate. $D_P^2(t)$ describes a "scroll" like page wherein K new content pieces are appended at a predictible, average rate. The two possible model fits are shown below.

```python
Image(filename='D_interp.png') 
```

A least squares method is used to find the optimal parameters to fit the Churn and Scroll models. The model with a smaller absolute error is chosen. In the above example, the churn model would be used as it better approximates the discrete function. It is worth noting that K itself cannot easily be fit via gradient descent as it takes on discrete, integer values. In this case it is valuable to fit several K>=2 models so long as the least squares error continues to decrease with increased K.

Now that a continuous approximation to our divergence has been formed, let's solve for the time, $t$, which solves the following integral equation:

$$
T = t * D_P(t) - \int_0^t{D_p(t')dt'}
$$

As a first step, rewrite the equation as the equivalent root-finding problem 

$$
0 = t * D_P(t) - \int_0^t{D_p(t')dt'} - T
$$

The roots (or zeros) of this equation can be found using the classic [Newton Raphson method](http://en.wikipedia.org/wiki/Newton's_method). For every gradient step in Newton Raphson, $D$ must be integrated. A simple [Gaussian quadrature](http://en.wikipedia.org/wiki/Gaussian_quadrature) method is used with relatively high error tolerance for fast evaluation. As we're integrating a smooth, exponential-like function, the integral should converge quickly. After integration, $U$ is calculated as $D_P(t)$ minus the integral. Finally, this page will be scheduled for a refresh at the time, $t$ solving $U_P(t)=T$.

### Online: caching paramteres, recalculating Poisson, etc.

It is worth periodically updating $D_P(t)$ to improve the fit of the above distributions. The optimal time for recalcuation is up for debate but computational resources aside, should be performed after every re-crawl of a page. A recalcuation is a low memory operation: simply re-average the discrete distribution, fit with a series of Poisson-like distributions and find the $t$ solving $U_P(t)=T$.

### Aside initial population of index
Initial population of the index should be performed via uniform page refreshes at a rate affordable given our utility. Again, debate remains as to the optimal time to transfer from a uniform refresh to the utility refresh described above. Ideally, error of the fit Poisson distribution should be tracked every time a new page is added. The error should descrease over time and a transition should occur after the error drops below a reasonable, global threshold $\epsilon$.
