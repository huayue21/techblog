# Google KV

One of the most exciting talks of the conference was a tutorial run by Evgeniy Gabrilovich (Google's) and Antoine Bordes (Facebook) on [Constructing and Mining Web-scale Knowledge Graphs](https://www.dropbox.com/s/inbuaxkzqsjbmv5/KDD14-T2-Bordes-Gabrilovich.pdf?dl=0). The two split the tutorial into two sessions. The first session, run by Evgeniy, provided an overview of [Freebase](http://www.freebase.com/), a publically available, structured knowledge graph of human information. Evgeniy delved into the uses of and challenges with maintaining such a large, diverse graph of knowledge. The second session, run by Antoine, focuses on a few, important problems and potential solutions faced by the Freebase community.

Evgeniy began with a beautiful overview of the purpose of Knowledge Vault: primairly, entity extraction. This begs the questions, what is an entity and why does it need to be extracted? 

To answer the first question, an entity is any sort of *structued* information that can be understood from *any* arbitary data source. We focus on one schema for representing structured information - Resource Description Framework (RDF) triples. RDFs are a representation of an atomic unit of human knowledge in a triple form: (subject, predicate value). For example, the sentence "Deadmau5 is an electronic artist" can be written in triple form (Deadmau5, is_a, electronic_artist). However, not all information is so easily mapped into a RDF triple. Take the sentence "URX is a tech startup based in South Park, San Francisco". There are a multitude of facts here. We could derive:

```
(URX, isa, start_up)
(URX, is_located, South_Park)
(South_Park, is_located, San_Francisco)
(San_Francisco, isa, location)
...
```

It starts to become apparent that extracting structured facts from unstructed text is a complex problem! To this end, the information retrieval community has been grateful for Freebase: the largest, freely availabe collection of RDF triples on human knowledge. Freebase was largely seeded from Wikipedia entries with supplemental information provided from [musicbrainz](musicbrainz.org) and publicly available records. All in all, Freebase contains approximately *637 million* facts (triples) and is growing everyday. 

One amazing consequence of represeting human knowledge in RDF triples is that this form readily maps to a graph structure. Subjects (and Values) are nodes in the graph with edges labeled as the predicate. Once a graph the size of Freebase is built, classic graph operations can quickly identify facts, discover relevant content to a given subject, and even interpret questions.

This leads to the second question: why are structured, graph entities useful? The simplest answer can be discovered with a [google search for the US Open](https://www.google.com/search?q=us+open).

Beyond boosting impoving search quality, Evgeniy alludes to several more applications of Freebase. The utility of Facebook's Graph Search can be boosted using Freebase. Searches such as "People who like Harvard University and Basketball at work at Facebook" is interpreted one fact at a time and matched against a combination of Facebook's internal data and Freebase's knowledge graph. Siri, Evi and other question answering intelligence are powered using Freebase. The learners inside these technologies attempt to match user questions with Freebase facts and return the facts as answers.


At the top, extraction of information is broken into two domains. The first, text extractors, answer the question what can we get from a web page? The second is Freebase priors; what is already known that can improve our ability to form facts?

Evgeniy mentions four distinct text extractors: NL text, Webmaster annotations, Page structure and Tables. For details on each process see the KDD talk and referenced papers. 

At a high level, NL text uses techniques learned from decades of natural language processing to tag words within sentences. Tags typically encompass both parts of speech and more specific information (Location, Event, Name, etc). The most popular of these models use [conditional random fields](http://en.wikipedia.org/wiki/Conditional_random_field) trained over huge amounts of labeled text data to garner an understanding of tags. However, even the best NLP methods contain enough noise that it's wise to supplement NLP with a few other features. 

Webmaster annotations leverage [Schema.org](schema.org), hand labeled DOM elements, to extract strucuted facts. Unfortunately few sites use Schema.org markup and the ones that do tend to be both sparse with their labeling and inaccurate. Nevertheless, if a site is religious about sticking to Schema.org's schema a set of high quality features can be extracted.

Page structure uses the hierarchy of DOM elements as a feature to learn strucuted information. Given domains may re-use specific DOM elements when referring to similar classes of facts. Again, relying solely on page structure is dangerous as HTML templates are dynamic. However, once a small amount of information can be correlated with a specific DOM path, subsequent information can be confidently extracted.

Finally, tables on pages act as a wealth of information. Wikipedia's entry on [Jupiter](http://en.wikipedia.org/wiki/Jupiter) contains a large fact table on the right hand side of the page. If a table can be accurately identified and the enclosed information confidently extracted, a large number of structued facts can be deduced. 

The theme of the above text features is that individually, they're all noisy. However, taken in combination, in the form of an ensemble learner the features come together to generate an accurate feature extractor. The same holds true for freebase priors. Again, it's beyond the scope of this post to dive into the principles of deep neural networks and path ranking algorithms. At a cursory level, neural networks gather *unstructured* information about human language from volumes of human knowledge. Path ranking algorithms differ by leveraging the inherent graph structure of freebase to boost fact finding.

As is common in data science, features gleaned from different domains are often fused via ensemble learners. GKV fuses extractors and priors using a third ensemble learner into the system they use today.
